```python
import pandas as pd
import numpy as np
```

## Import The Data

### Print out shape and first 5 lines

```python
df = pd.read_csv('ds-job-listings.csv')
print(df.shape)
df.head()
```

## Clean the Data

```python
# Drop Unnamed: 0 column
df = df.drop(['Unnamed: 0'], axis=1)

# Remove listings where Description or Link == None
df = df[df.Link != "None"]
df = df[df.Description != "None"]

# Remove Duplicates
df = df.drop_duplicates()
print(df.shape)
df.head()
```

## Remove Rows where Title does not contain "Data Scientist"

```python
DS = df[df.Title.str.contains("Data Scientist") == True]
print(DS.shape)
DS.head()
```

```python
import gensim
import nltk
from nltk.corpus import stopwords 
nltk.download('stopwords')
import re

def transformText(text):
    
    stops = set(stopwords.words("english"))
    
    # Convert text to lower
    text = text.lower()
    # Removing non ASCII chars    
    text = re.sub(r'[^\x00-\x7f]',r' ',text)
    
    # Strip multiple whitespaces
    text = gensim.corpora.textcorpus.strip_multiple_whitespaces(text)
    
    # Removing all the stopwords
    filtered_words = [word for word in text.split() if word not in stops]
    
    # Removing all the tokens with lesser than 3 characters
#     filtered_words = gensim.corpora.textcorpus.remove_short(filtered_words, minsize=3)
    
    # Preprocessed text after stop words removal
    text = " ".join(filtered_words)
    
    # Remove the punctuation
    text = gensim.parsing.preprocessing.strip_punctuation2(text)
    
    # Strip all the numerics
    text = gensim.parsing.preprocessing.strip_numeric(text)
    
    # Strip multiple whitespaces
    return gensim.corpora.textcorpus.strip_multiple_whitespaces(text)
#     return list(gensim.utils.tokenize(gensim.corpora.textcorpus.strip_multiple_whitespaces(text)))
    
    # Stemming
#     return gensim.parsing.preprocessing.stem_text(text)
```

```python
pd.options.display.max_colwidth = 500

DS['Description'] = DS['Description'].map(transformText)

DS.head()

DS.to_csv("DS.csv")
```

```python
pd.set_option('display.max_columns', 500)
from sklearn.feature_extraction.text import CountVectorizer

docs = DS['Description'].tolist()
vect = CountVectorizer(min_df=.01, max_df=1.0, binary=True, ngram_range = (1,3), stop_words = 'english')
X = vect.fit_transform(docs)
doc_matrix = pd.DataFrame(X.A, columns=vect.get_feature_names())

doc_matrix.head()
```

```python
data_final = DS.join(doc_matrix)

print(data_final.shape)
data_final.head()
```

```python
sum_dict = {}
columns = list(data_final.columns)
columns.remove('Title')
columns.remove('Location')
columns.remove('Company')
columns.remove('Synopsis')
columns.remove('Description')
columns.remove('Link')
columns.remove('Search')

for column in columns:
    header = str(column)
    sum_dict[header] = data_final[header].sum()
    
print(len(sum_dict))
print('\n "First 100 items: ')
print(list(sum_dict.keys())[:100])
```

```python
print('Sort by items:')
for key, value in sorted(sum_dict.items(), key=lambda item: (item[1], item[0])):
    print ("%s: %s" % (key, value))
```
